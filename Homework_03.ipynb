#### Задание 3:
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
def circle(A=[0, 0], R=5):
    a, b = A
    x = np.linspace((a - R), (a + R), 121)
    y1 = np.sqrt(np.power(R, 2) - np.power((x - a), 2)) + b
    y2 = -1 * y1
    plt.plot(x, y1, color='red')
    plt.plot(x, y2, color='red')
    plt.xlim([(a - R - 1), (a + R + 1)])
    plt.ylim([(b - R - 1), (b + R + 1)])
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
circle()
```

```python
def ellipse(A=[0, 0], a=10, b=5):
    x0, y0 = A
    x = np.linspace((x0 - a), (x0 + a), 121)
    y1 = np.sqrt(np.power(b, 2) - np.power(b, 2) *  np.power((x - x0), 2) / np.power(a, 2)) + y0
    y2 = -1 * y1
    plt.plot(x, y1, color='green')
    plt.plot(x, y2, color='green')
    plt.xlim([(x0 - a - 1), (x0 + a + 1)])
    plt.ylim([(y0 - b - 1), (y0 + b + 1)])
    plt.show()
```

```python
ellipse()
```

```python
def hyperbola(x=np.linspace(-5, 5, 121), a=1, b=1):
    y1 = np.sqrt((x * b / a - b) * ((x * b / a + b)))
    y2 = -1 * y1
    plt.plot(x, y1, color='yellow')
    plt.plot(x, y2, color='yellow')
    plt.show()
```

```python
hyperbola()
```

#### Задание 2:
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

```python
x, y = 2, 5
a = np.pi / 4 
sin_a = np.sin(a)
cos_a = np.cos(a)
x1, y1 = x*sin_a + y*cos_a, x*-sin_a + y*cos_a
plt.plot(x, y, color='green', marker='o')
plt.plot(x1, y1, color='red', marker='o')
```

```python
norm = np.round(np.linalg.norm([x, y]), 4)
norm1 = np.round(np.linalg.norm([x1, y1]), 4)
print(norm, norm1)
```

```python
print(norm == norm1)
```

При линейном преобразовании расстояние между точками сохранилось. Докажем, что преобразование является ортогональным.

Условия ортогональности линейного преобразования:

* a11^2 + a12^2 = 1
* a21^2 + a22^2 = 1
* a11a12 + a21a22 = 0
* a11a22 - a12a21 != 0

```python
a11, a12, a21, a22 = sin_a, cos_a, -sin_a, cos_a
print(f'a11^2 + a12^2 = 1: {np.round(a11**2 + a12**2, 1) == 1.0}')
print(f'a21^2 + a22^2 = 1: {np.round(a21**2 + a22**2, 1) == 1.0}')
print(f'a11*a12 + a21*a22 = 0: {np.round(a11*a12 + a21*a22, 1) == 0.0}')
print(f'a11*a12 + a21*a22 = 0: {np.round(a11*a22 - a12*a21, 1) != 0.0}')
```

Из вычислений выше, мы видим, что наше линейное преобразование точек является ортогональным. Так как расстояние между точками сохранились, то мы можем сделать вывод, что при ортогональным преобразовании расстояние между точками сохраняется.

```python
def polar_to_cartesian(R, a):
    x = R * np.cos(np.radians(a))
    y = R * np.sin(np.radians(a))
    return x, y
```

```python
def polar_circle(R=1):
    phi = np.linspace(0, 2 * np.pi, 201)
    r = np.linspace(R, R, 201)
    plt.polar(phi, r, color='blue')
    plt.show()
```

#### Задание 4:
1. Решите систему уравнений:

$y = x^2 – 1$

$exp(x) + x∙(1 – y) = 1$

```python
#для начала построим линии этих уравнений - чтобы увидеть точки пересечения, тем самым мы увидим где
#находятся корни уравнения на графике
x = np.linspace(-5, 5, 201)
plt.plot(x, x**2-1)
plt.plot(x, (np.exp(x)+x-1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equitations(p):
    x, y = p
    return y - x**2 + 1, np.exp(x) + x*(1-y) - 1
```

```python
x1, y1 = fsolve(equitations, (-1, 0))
print(x1, y1)
```

```python
x2, y2 = fsolve(equitations, (2, 5))
print(x2, y2)
```

```python
x3, y3 = fsolve(equitations, (4, 15))
print(x3, y3)
```

```python
print(f"Корнями системы уравнений являются точки с координатами: {x1, y1}, {x2, y2}, {x3, y3}")
```
